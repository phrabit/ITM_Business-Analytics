```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Excel 파일 읽기
df = pd.read_excel('/content/drive/MyDrive/B A 데이터 모음/00클러스터링/Preprocessed_datasets/Income_Pdensity_Company.xlsx')

df
```
