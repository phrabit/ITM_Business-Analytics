```python
import pandas as pd
import numpy as np
```

Name the columns

```python
column_names = [
    'zone_origin', 'code_origin', 'node_origin', 'zone_dest', 'code_dest', 'node_dest',
    'walk_and_bike', 'truck', 'other_bus', 'train', 'car', 'taxi', 'bus', 'subway', 'bus_and_subway'
    ]
```

## Import the original data<br> (국가 교통 DB)

Drop the rows whose origins are not in Seoul (not import Zone number >= 425).

```python
# Specify the number of rows to read
num_rows_to_read = 554167

# Read the CSV file up to the specified number of rows
od_am_peak = pd.read_csv('/content/drive/MyDrive/B A 데이터 모음/00클러스터링/통행량_수단전체/4. OD/OD_MMODE_19_AMPEAK_F_230425.csv', header=None, nrows=num_rows_to_read)
od_pm_peak = pd.read_csv('/content/drive/MyDrive/B A 데이터 모음/00클러스터링/통행량_수단전체/4. OD/OD_MMODE_19_PMPEAK_F_230425.csv', header=None, nrows=num_rows_to_read)
od_full = pd.read_csv('/content/drive/MyDrive/B A 데이터 모음/00클러스터링/통행량_수단전체/4. OD/txt_F.TXT', header=None, nrows=num_rows_to_read)

od_am_peak.columns = column_names
od_pm_peak .columns = column_names
od_full .columns = column_names
```

```python
od_am_peak
```

Drop the rows whose destinations are not in Seoul (drop Zone number >=425).

```python
# Create a boolean mask for rows where 'zone_dest' is larger than 425
mask_am_peak = od_am_peak['zone_dest'] >= 425
mask_pm_peak = od_pm_peak['zone_dest'] >= 425
mask_full = od_full['zone_dest'] >= 425

# Use the mask to drop rows that meet the condition
od_am_peak = od_am_peak[~mask_am_peak]
od_pm_peak = od_pm_peak[~mask_pm_peak]
od_full = od_full[~mask_full]
```

```python
od_am_peak.shape
```

Drop the columns:  'walk_and_bike', 'truck', 'other_bus', 'train', 'bus_and_subway'

```python
od_am_peak = od_am_peak.drop(columns=['zone_origin', 'node_origin', 'zone_dest', 'node_dest', 'walk_and_bike', 'truck', 'other_bus', 'train'])
od_pm_peak = od_pm_peak.drop(columns=['zone_origin', 'node_origin', 'zone_dest', 'node_dest', 'walk_and_bike', 'truck', 'other_bus', 'train'])
od_full = od_full.drop(columns=['zone_origin', 'node_origin', 'zone_dest', 'node_dest', 'walk_and_bike', 'truck', 'other_bus', 'train'])
```

```python
od_am_peak.shape
```

```python
od_am_peak['transportation_total'] = od_am_peak[['car', 'taxi', 'bus', 'subway', 'bus_and_subway']].sum(axis=1)
od_pm_peak['transportation_total'] = od_pm_peak[['car', 'taxi', 'bus', 'subway', 'bus_and_subway']].sum(axis=1)
od_full['transportation_total'] = od_full[['car', 'taxi', 'bus', 'subway', 'bus_and_subway']].sum(axis=1)
```
