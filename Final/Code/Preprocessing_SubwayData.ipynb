```python
import pandas as pd
```

```python
df = pd.read_csv("subway2019.csv")

df
```

```python
# Remove the "등록일자" column
df = df.drop('등록일자', axis=1)
df
```

```python
columns_to_remove = ['사용일자', '노선명']
df = df.drop(columns=columns_to_remove)
df
```

```python
# Group by '역명' and aggregate the sum of '승차총승객수' and '하차총승객수'
aggregated_df = df.groupby('역명').agg({
    '승차총승객수': 'sum',
    '하차총승객수': 'sum'
}).reset_index()

# Create a new column '총승객수' by summing '승차총승객수' and '하차총승객수'
aggregated_df['총승객수'] = aggregated_df['승차총승객수'] + aggregated_df['하차총승객수']

# Drop duplicates based on the '역명' column
result_df = aggregated_df.drop_duplicates(subset='역명', keep='first')

result_df
```
